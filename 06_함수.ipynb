### âœ… ì¬ê·€í•¨ìˆ˜(recursive)
: í•¨ìˆ˜ ë‚´ë¶€ì—ì„œ ìê¸° ìì‹ ì„ í˜¸ì¶œí•˜ëŠ” í•¨ìˆ˜

#### âœ”ï¸ íŒ©í† ë¦¬ì–¼
- n! = 1 * 2 * 3 * ... * n

````python
# ë°˜ë³µë¬¸
def fact(n):
    result = 1

    while n > 1:
        result = result * n
        n = n - 1
    return result

print(fact(5))
````

````python
# ì¬ê·€
def factorial(n):
    if n <= 1:
        return 1
    else:
        return factorial(n-1) * n

print(factorial(5))
````

#### âœ”ï¸ í”¼ë³´ë‚˜ì¹˜ ìˆ˜ì—´
- F(0) = F(1) = 1
- F(N) = F(N-1) + F(N-2)

````python
# ë°˜ë³µë¬¸
def fib_loop(n):
    result = [1, 1]

    for i in range(1, n):
        end1 = result[-1]
        end2 = result[-2]

        fib_num = end1 + end2

        result.append(fib_num)

    return result[-1]

print(fib_loop(10))
````

````python
# ì¬ê·€
def fib_rec(n):
    if n == 0 or n == 1:
        return 1
    else:
        return fib_rec(n-1) + fib_rec(n-2)

print(fib_rec(10))
````
